```python
corpus ="""Hello welcome, to my NLP tuetorial practice.
Please do follow me for this samples! you difinitely like this NLP example. It's Super Excited!!

"""
```

```python
print(corpus)
```

```python
## Tokenization
## Sentences --> Paragraphs
from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Next Tokenization
## Paragraph --> words
## Sentence--> Workds
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
## Another type of tokenizer
## It splits text into alphabetic/numeric characters and everything else. It will aggressively separate every single piece of punctuation into its own token.
## Example: I'm $\rightarrow$ ['I', "'", 'm']
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
## TreebankWordTokenizer
## Key Feature: It is famous for how it handles contractions. It splits "don't" into "do" and "n't".
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
