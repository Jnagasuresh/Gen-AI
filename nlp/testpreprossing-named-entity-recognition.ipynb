```python
sentence="""The Eiffel Tower was built from 1887 to 1889 by French engineer Gustave Eiffel, whose company
speciliased in building metal frameworks and structures."""
```

```python
import nltk
words=nltk.word_tokenize(sentence)
```

```python
nltk.download('maxent_ne_chunker')
```

```python
nltk.download('words')
```

```python
tag_elements=nltk.pos_tag(words)
```

```python
tag_elements
```

```python
nltk.ne_chunk(tag_elements).draw()
```
