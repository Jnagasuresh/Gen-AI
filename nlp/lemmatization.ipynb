# Lemmatization

## Wordnet Lemmatizer
Lemmatization technique is like stemming. the output we will get after lemmatization is called 'lemma', which is na root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same thing.

NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus. this class uses morphy() function to the WordNet CorpusReader class to find a lemma.

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
'''
POS- Noun-n
verb-v
adjective-a
adverb-r
'''
lemmatizer.lemmatize("going",pos='v')
```

```python
words=["eating", "eats","eaten","writing","writes","programming","programs","history","finally","finalize"]
```

```python
for word in words:
    print(word+"---> "+lemmatizer.lemmatize(word,pos='v'))
```

```python
lemmatizer.lemmatize("goes",pos='v')
```

```python
lemmatizer.lemmatize('fairly'),lemmatizer.lemmatize('sportingly')
```
