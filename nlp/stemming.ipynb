# Stemming
Stemming is a process of reducing a word to its root form. It reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of wordss known as lemma. For example, the root form of "running", "ran", and "runs" is "run".

```python
## Classification Problem
## Comments of product is a possitive review or negative review
## Reviews --> Eating, Eat, Eaten --> [going, gone, goes -> go]
words=["eating", "eats","eaten","writing","writes","programming","programs","history","finally","finalize"]
```

# PorterStemmer

```python
### PorterStemmer
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in words:
    print(word+"-->  "+ stemming.stem(word))
    
```

```python
stemming.stem('congratulations')
```

```python
stemming.stem('sitting')
```

# RegexpStemmer Class
NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffic that matches the expression. 

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|abl$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

# Snowball Stemmer


```python
from nltk.stem import SnowballStemmer
```

```python
snowball_stemmer= SnowballStemmer('english')
```

```python
for word in words:
    print(word+"-->  "+ snowball_stemmer.stem(word))
```
